```python
dir_prefix = 'drive/My Drive/Colab Notebooks/Diplomski'
```

```python
underlying_data_path = os.path.join(dir_prefix, 'multi_label_dataset', 'tooth_info.csv')
orig_df = pd.read_csv(underlying_data_path, sep = ";")

corrections_csv_path = os.path.join(dir_prefix, 'corrections_csvs', 'all_corrections_THRES-0.9.csv')
corrections = pd.read_csv(corrections_csv_path, sep=';')

for i,r in enumerate(corrections.iterrows()):
  row = r[1]
  i_nm = row['img_name']
  t_id = row['tooth_id']
  orig_df.loc[(orig_df['img_name'] == i_nm) & (orig_df['tooth_id'] == t_id), ['bridge', 'endofilling', 'filling', 'crown']] = row[['bridge', 'endofilling', 'filling', 'crown']].values


corrected_data_path = os.path.join(dir_prefix, 'multi_label_dataset', 'corrected_tooth_info.csv')
orig_df.to_csv(corrected_data_path, sep=';', index=False)
```
